```python
import pandas as pd
df_apr_may = pd.read_csv("Classes April-May 2018.csv")
df_june = pd.read_csv("Classes June 2018.csv")
print("April-May shape: ", df_apr_may.shape)
print("June shape: ", df_june.shape)
df_apr_may.head(), df_june.head()
    
```

```python
df = pd.concat([df_apr_may, df_june], ignore_index=True)
print("Combined shape:", df.shape)
df.head()
```

```python
df = df.drop_duplicates()
print("After dropping duplicates:", df.shape)
```

```python
for col in df.select_dtypes(include=["object"]).columns:
    df[col] = df[col].astype(str).str.strip()

df.head()
```
