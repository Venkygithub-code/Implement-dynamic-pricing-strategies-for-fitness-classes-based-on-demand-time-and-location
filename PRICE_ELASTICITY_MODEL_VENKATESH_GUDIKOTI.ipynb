```python
import pandas as pd

df = pd.read_csv("Fitness_Classes_Cleaned_Preprocessed.csv")

df.columns = ( 
    df.columns
    .str.lower()
    .str.strip()
    .str.replace(" ", "_" , regex=False)
    .str.replace("(", "" , regex=False)
    .str.replace(")", "" , regex=False)
)

df = df[["price_inr", "number_booked"]]
df.columns = ["price", "attendance"]

df.head()
```

```python
import numpy as np

df=df[(df["price"] > 0) & (df["attendance"] > 0)]

df["log_price"] = np.log(df["price"])
df["log_attendance"] = np.log(df["attendance"])


```

```python
import statsmodels.api as sm

x=sm.add_constant(df["log_price"])
y=df["log_attendance"]

model = sm.OLS(y, x).fit()
print(model.summary())
```

```python
elesticity = model.params["log_price"]
elesticity
```

```python
from sklearn.metrics import mean_squared_error
predictions = model.predict(x)
mse = mean_squared_error(y, predictions)
mse
```
